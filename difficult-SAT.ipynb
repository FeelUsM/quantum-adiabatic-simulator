```python
import random
from random import randint
import pycosat
```

```python
def cnf_random_generator(n_variables,m_clauses):
    cnf=[]
    for i in range(m_clauses):
        a=sorted(random.sample(range(1,n_variables+1),3))
        b=[(-1)**randint(0,1),(-1)**randint(0,1),(-1)**randint(0,1)]
        c=[a[i]*b[i] for i in range(3)]
        cnf.append(c)
    return cnf
```

```python
cnf = sorted(cnf_random_generator(3,10),key=lambda l:[abs(l[0]),abs(l[1]),abs(l[2])])
for i in range(len(cnf)):
    print(cnf[i],',')
pycosat.solve(cnf)
```

```python
def SAT_2_NAE3SAT(SAT):
    z = max([max([abs(lit) for lit in clause]) for clause in SAT])+1
    NAE4SAT = [clause+[z] for clause in SAT]
    NAE3SAT = []
    for i in range(len(NAE4SAT)):
        cl = NAE4SAT[i]
        NAE3SAT.append([cl[0], cl[1],   z+1+i ])
        NAE3SAT.append([cl[2], cl[3], -(z+1+i)])
    return z,NAE3SAT
```

```python
SAT_2_NAE3SAT(cnf)
```

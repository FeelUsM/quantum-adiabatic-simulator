```python
import numpy as np
import random
from random import randint
import pycosat
```

```python
def cnf_random_generator(n_variables,m_clauses):
    cnf=[]
    for i in range(m_clauses):
        a=sorted(random.sample(range(1,n_variables+1),3))
        b=[(-1)**randint(0,1),(-1)**randint(0,1),(-1)**randint(0,1)]
        c=[a[i]*b[i] for i in range(3)]
        cnf.append(c)
    return cnf
```

```python
cnf = sorted(cnf_random_generator(20,102),key=lambda l:[abs(l[0]),abs(l[1]),abs(l[2])])
for i in range(len(cnf)):
    print(cnf[i],',')
pycosat.solve(cnf)
```
